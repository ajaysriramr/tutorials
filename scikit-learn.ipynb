```python
#sklearn datasets
from sklearn import datasets

iris = datasets.load_iris()
y = iris.target
X = iris.data

# for i in range(len(X[0])):
#     plt.figure()
#     plt.scatter(X[:,i],y)
```

```python
#train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#KNN

from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

knn = neighbors.KNeighborsClassifier(n_neighbors=15,weights='uniform')
knn.fit(X_train,y_train)
predictions = knn.predict(X_test)
accuracy = metrics.accuracy_score(y_test,predictions)
accuracy
```

```python
#SVM

from sklearn import svm

model = svm.SVC()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = metrics.accuracy_score(y_test,predictions)
accuracy
```

```python
# Linear Regression

from sklearn import linear_model

boston = datasets.load_boston()
X = boston.data
y = boston.target

l_reg = linear_model.LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

model = l_reg.fit(X_train,y_train)
predictions = model.predict(X_test)
```

```python
#KMeans

from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import scale

bc = load_breast_cancer()

Xb = bc.data
yb = bc.target

Xb_train, Xb_test, yb_train, yb_test = train_test_split(Xb, yb, test_size = 0.2)

model = KMeans(n_clusters = 2, random_state = 0)
model.fit(Xb_train)
predictions = model.predict(Xb_test)
label = model.labels_
accuracy = accuracy_score(yb_test, predictions)
print(accuracy)
```
